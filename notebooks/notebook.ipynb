## top 10 handsets used by customers

```python
# SQL query to find the top 10 handsets used by customers
# This query groups the data by Handset Manufacturer and Handset Type,
# counts the number of occurrences for each combination, and orders the results
# by the count in descending order, limiting the output to the top 10.
sql_query = """
SELECT "Handset Manufacturer", "Handset Type", COUNT(*) AS "Count"
FROM "public"."xdr_data"
GROUP BY "Handset Manufacturer", "Handset Type"
ORDER BY "Count" DESC
LIMIT 10;
"""

# Execute the SQL query and store the result in a pandas DataFrame
df = pd.read_sql(sql_query, con=engine)

# Display the DataFrame containing the top 10 handsets used by customers
print(df)
```

```python
# Set the size of the figure
plt.figure(figsize=(10, 6))

# Create a horizontal bar plot
plt.barh(df['Handset Manufacturer'] + ' ' + df['Handset Type'], df['Count'], color='skyblue')

# Set the title and axis labels for the plot
plt.title('Top 10 Handsets Used by Customers')
plt.xlabel('Count')
plt.ylabel('Handset Manufacturer and Type')

# Rotate the y-axis labels if needed
plt.yticks(rotation=0)

# Display the plot
plt.show()
```

```python
# Group the data by 'Handset Manufacturer' and count the number of occurrences
manufacturer_counts = df['Handset Manufacturer'].value_counts()

# Get the top 3 handset manufacturers
top_manufacturers = manufacturer_counts.head(3)

# Display the top 3 handset manufacturers
print(top_manufacturers)
```

```python
# Plot the top 3 manufacturers as a bar plot
plt.bar(top_manufacturers.index, top_manufacturers.values)

# Customize the plot
plt.xlabel('Handset Manufacturer')
plt.ylabel('Count')
plt.title('Top 3 Handset Manufacturers')

# Rotate x-axis labels
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
