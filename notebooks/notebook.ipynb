## top 10 handsets used by customers

```python
# SQL query to find the top 10 handsets used by customers
# This query groups the data by Handset Manufacturer and Handset Type, removing null values,
# counts the number of occurrences for each combination, and orders the results
# by the count in descending order.


# SQL query to retrieve data for all rows with non-null "Handset Manufacturer"
sql_query = """
SELECT "Handset Manufacturer", "Handset Type", COUNT(*) AS "Count"
FROM "public"."xdr_data"
WHERE "Handset Manufacturer" IS NOT NULL
GROUP BY "Handset Manufacturer", "Handset Type"
ORDER BY "Count" DESC
LIMIT 10;
"""

# Execute the SQL query and store the result in a pandas DataFrame
df_top10_handsets = pd.read_sql(sql_query, con=engine)

# Display the DataFrame containing the handsets used by customers (all rows)
print(df_top10_handsets)
```

```python
# Set the size of the figure
plt.figure(figsize=(10, 6))

# Create a horizontal bar plot
plt.barh(df_top10_handsets['Handset Manufacturer'] + ' ' + df_top10_handsets['Handset Type'], df_top10_handsets['Count'], color='skyblue')

# Set the title and axis labels for the plot
plt.title('Top 10 Handsets Used by Customers')
plt.xlabel('Count')
plt.ylabel('Handset Manufacturer and Type')

# Rotate the y-axis labels if needed
plt.yticks(rotation=0)

# Display the plot
plt.show()
```

```python
# Define the SQL query to select the top 3 handset manufacturers, excluding null values
sql_query = """
SELECT "Handset Manufacturer", COUNT(*) AS count
FROM "public"."xdr_data"
WHERE "Handset Manufacturer" IS NOT NULL
GROUP BY "Handset Manufacturer"
ORDER BY count DESC
LIMIT 3
"""

# Execute the SQL query and store the result in a pandas DataFrame
df_top3_handset_manufacturers = pd.read_sql(sql_query, con=engine)

# Display the top 3 handset manufacturers
print(df_top3_handset_manufacturers)
```

```python
# Plot the top 3 manufacturers as a bar plot
plt.bar(df_top3_handset_manufacturers['Handset Manufacturer'], df_top3_handset_manufacturers['count'])

# Customize the plot
plt.xlabel('Handset Manufacturer')
plt.ylabel('Count')
plt.title('Top 3 Handset Manufacturers')

# Rotate x-axis labels if necessary
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

## Top 5 handsets for each of the top 3 manufacturers

```python
conn = engine.raw_connection()
cur = conn.cursor()

# Step 1: Get the top 3 manufacturers
cur.execute("""
SELECT "Handset Manufacturer", COUNT(*) AS count
FROM "public"."xdr_data"
WHERE "Handset Manufacturer" IS NOT NULL
GROUP BY "Handset Manufacturer"
ORDER BY count DESC
LIMIT 3
""")

top_manufacturers = cur.fetchall()

# Step 2: For each manufacturer, get the top 5 handsets
for manufacturer in top_manufacturers:
    manufacturer_name = manufacturer[0]
    cur.execute(f"""
    SELECT "Handset Type", COUNT(*) AS count
    FROM "public"."xdr_data"
    WHERE "Handset Manufacturer" = '{manufacturer_name}'
    GROUP BY "Handset Type"
    ORDER BY count DESC
    LIMIT 5
    """)
    top_handsets = cur.fetchall()
    print(f"Top 5 handsets for {manufacturer_name}:")
    for handset in top_handsets:
        print(f"{handset[0]}: {handset[1]}")
```

```python
# Prepare data for plotting
manufacturer_names = [manufacturer[0] for manufacturer in top_manufacturers]

for manufacturer in top_manufacturers:
    manufacturer_name = manufacturer[0]
    cur.execute(f"""
    SELECT "Handset Type", COUNT(*) AS count
    FROM "public"."xdr_data"
    WHERE "Handset Manufacturer" = '{manufacturer_name}'
    GROUP BY "Handset Type"
    ORDER BY count DESC
    LIMIT 5
    """)
    top_handsets = cur.fetchall()

    handsets = [handset[0] for handset in top_handsets]
    counts = [handset[1] for handset in top_handsets]

    # Plotting the data
    plt.bar(handsets, counts)
    plt.title(f"Top 5 Handsets for {manufacturer_name}")
    plt.xlabel("Handset Type")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()

# Close the cursor and connection
cur.close()
conn.close()
```
